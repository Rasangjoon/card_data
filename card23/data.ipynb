```python
import pandas as pd
import re

# CSV 파일을 데이터프레임으로 불러오기
df = pd.read_csv('card_data.csv')

# 추가항목 6칼럼에서 할인율 정보 추출하기
def extract_discount_rate(text):
    if isinstance(text, float):
        return None
    # 정규표현식을 사용하여 할인율 추출 (7% 할인 또는 1% 적립)
    pattern = r'(\d+)%'
    match = re.search(pattern, text)
    if match:
        return int(match.group(1))
    else:
        return None

# '추가항목6칼럼'에서 할인율 추출하여 새로운 열 '할인율(%)' 추가
df['추가항목6칼럼'] = df['추가항목6'].astype(str)
df['할인율(%)'] = df['추가항목6'].apply(extract_discount_rate)

# 새로운 데이터프레임 생성
analyzed_df = df[['카드이름', '할인율(%)']]  # '카드사' 열과 '할인율(%)' 열을 선택하여 새로운 데이터프레임 생성

# 결과 확인
print(analyzed_df.head())
```

```python
import pandas as pd
import re

# CSV 파일을 데이터프레임으로 불러오기
df = pd.read_csv('card_data.csv')

# 추가항목 6칼럼에서 쿠팡·11번가 등 연최대 6만원할인쇼핑 내용과 할인율 추출하기
def extract_discount_info(text):
    if isinstance(text, float):
        return None, None
    # 정규표현식을 사용하여 쿠팡·11번가 등 연최대 6만원할인쇼핑 내용 추출
    pattern = r'(.+) (\d+)%'
    match = re.search(pattern, text)
    if match:
        item_name = match.group(1)
        discount_rate = int(match.group(2))
        return item_name, discount_rate
    else:
        return None, None

# '추가항목6칼럼'에서 쿠팡·11번가 등 연최대 6만원할인쇼핑 내용과 할인율 추출하여 새로운 열 추가
df['추가항목2'] = df['추가항목2'].astype(str)
df['할인내용2'], df['할인율2(%)'] = zip(*df['추가항목2'].apply(extract_discount_info))


df['추가항목3'] = df['추가항목3'].astype(str)
df['할인내용3'], df['할인율3(%)'] = zip(*df['추가항목3'].apply(extract_discount_info))

df['추가항목4'] = df['추가항목4'].astype(str)
df['할인내용4'], df['할인율4(%)'] = zip(*df['추가항목4'].apply(extract_discount_info))

df['추가항목5'] = df['추가항목5'].astype(str)
df['할인내용5'], df['할인율5(%)'] = zip(*df['추가항목5'].apply(extract_discount_info))

df['추가항목6'] = df['추가항목6'].astype(str)
df['할인내용6'], df['할인율6(%)'] = zip(*df['추가항목6'].apply(extract_discount_info))

df['추가항목7'] = df['추가항목7'].astype(str)
df['할인내용7'], df['할인율7(%)'] = zip(*df['추가항목7'].apply(extract_discount_info))

df['추가항목8'] = df['추가항목8'].astype(str)
df['할인내용8'], df['할인율8(%)'] = zip(*df['추가항목8'].apply(extract_discount_info))

df['추가항목9'] = df['추가항목9'].astype(str)
df['할인내용9'], df['할인율9(%)'] = zip(*df['추가항목2'].apply(extract_discount_info))

# 필요한 열만 선택하여 새로운 데이터프레임 생성
analyzed_df = df[['카드이름', '할인내용2', '할인율2(%)', '할인내용3', '할인율3(%)','할인내용4', '할인율4(%)',
                  '할인내용5', '할인율5(%)','할인내용6', '할인율6(%)', '할인내용7', '할인율7(%)','할인내용8', '할인율8(%)',
                  '할인내용9', '할인율9(%)']]

# 결과 확인
print(analyzed_df.head())
```
