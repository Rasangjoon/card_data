```python
import requests
import json
import csv
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}

URL = 'https://card-search.naver.com/graphql'

composition_list = []

def crawl_data(cardAdId):
    params = {
        "operationName": "smartSearch",
        "variables": {
            'basePayment': 0,
            'bizType': 'CPC',
            'device': 'pc',
            'isRefetch': False,
            'maxAnnualFee': 0,
            'minAnnualFee': 0,
            'pageNo': 1,
            'pageSize': 10,
            'sortMethod': 'ri',
            'cardAdIds': [int(cardAdId.split('=')[1])]  # Extract cardAdId value from the URL
        },
        "query": '''query smartSearch($basePayment: Int, $bizType: BizType, $device: AdDeviceType, $isRefetch: Boolean, $maxAnnualFee: Int, $minAnnualFee: Int, $pageNo: Int, $pageSize: Int, $sortMethod: SortMethod, $cardAdIds: [Int]) {
            cardAdList(
                basePayment: $basePayment,
                bizType: $bizType,
                device: $device,
                isRefetch: $isRefetch,
                maxAnnualFee: $maxAnnualFee,
                minAnnualFee: $minAnnualFee,
                pageNo: $pageNo,
                pageSize: $pageSize,
                sortMethod: $sortMethod,
                cardAdIds: $cardAdIds
            ) {
                cardAds {
                    cardAdId
                    # Add other fields you want to extract here
                }
                totalSize
            }
        }'''
    }

    response = requests.post(URL, json=params, headers=headers)

    json_result = response.json()

    for data in json_result['data']['cardAdList']['cardAds']:
        composition_list.append({
            '사이트': data['cardAdId']
        })

crawl_data('cardAdId=1692')
# Add other cardAdIds here if needed

# Save the data to a CSV file
csv_file = 'composition_list.csv'
fieldnames = ['사이트']

with open(csv_file, mode='w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header
    writer.writeheader()

    # Write the data rows
    for row in composition_list:
        writer.writerow(row)

print(f'Data has been saved to {csv_file}.')
```

```python
import requests

def fetch_cards_data():
    url = 'https://card-search.naver.com/graphql'
    headers = {
        'Content-Type': 'application/json',
    }
    
    # GraphQL 쿼리를 작성합니다.
    query = '''
    query {
        cards {
            edges {
                node {
                    cardName
                }
            }
        }
    }
    '''
    
    # GraphQL 쿼리를 POST 요청으로 보냅니다.
    response = requests.post(url, headers=headers, json={'query': query})
    
    if response.status_code == 200:
        data = response.json()
        cards_data = data.get('data', {}).get('cards', {}).get('edges', [])
        card_names = [card['node']['cardName'] for card in cards_data]
        return card_names
    else:
        print('Failed to fetch data from the GraphQL endpoint.')
        return []

if __name__ == '__main__':
    card_names = fetch_cards_data()
    for name in card_names:
        print(name)
```
